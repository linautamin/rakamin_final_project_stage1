```python
#Univariate Analysis-boxplot
features = numerical
plt.figure(figsize=(12, 5))
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.boxplot(y=df[features[i]], color='green', orient='v')
    plt.tight_layout()
```

```python
plt.figure(figsize=(12, 5))
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.distplot(df[features[i]], color='gray')
    plt.tight_layout()
```

```python
plt.figure(figsize=(12, 5))
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.violinplot(y=df[features[i]], color='gray', orient='v')
    plt.tight_layout()
```

```python
#Univariate Analysis-boxplot
features = categorical
plt.figure(figsize=(12, 5))
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.countplot(x=df[features[i]], color='green')
    plt.tight_layout()
```
